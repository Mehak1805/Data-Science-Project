# Step 2: Data Cleaning & Preparation

## 1.Identifying and Analyzing Missing Values

## We next check for missing values in each column, because incomplete data can lead to incorrect analysis and errors during model building. Identifying where data is missing helps us decide whether to impute, clean, or remove those records.

```python
missing = df_data.isnull().sum().sort_values(ascending=False)
missing[missing > 0]
```

## 2.Checking for Duplicate Records
## We check for duplicate rows in the dataset because repeated entries can distort analysis, inflate counts, and bias statistical results. Identifying duplicates allows us to ensure that each patient visit is represented only once, improving the overall data quality and reliability.

```python
print("duplicates:", df_data.duplicated().sum())
```

## 3.Removing Duplicate Records
## After identifying duplicate entries, we remove them to ensure that each patient record appears only once in the dataset. Duplicate rows can mislead analysis by repeating the same information multiple times, affecting statistics, visualizations, and model training. Dropping duplicates improves data accuracy and ensures cleaner, more reliable results.

```python
df_data = df_data.drop_duplicates()
```

```python
df_data.head(40)
```

## 4.Cleaning Column Names
## We remove extra spaces from column names to keep them consistent and easy to work with. Clean column names help prevent referencing errors during analysis and preprocessing.

```python
# Cleaning column names (remove spaces)
df_data.columns = [col.strip() for col in df_data.columns]
df_data.head(40)
```

## 5.Standardizing Column Names
## We clean and standardize column names by removing spaces, converting them to lowercase, and replacing special characters with underscores.

```python
df_data.columns = (
    df_data.columns
    .str.strip()         # remove spaces at start/end
    .str.lower()         # convert to lowercase
    .str.replace(' ', '_')
    .str.replace('/', '_')
    .str.replace('-', '_')
)
print("Cleaned column names:")
print(df_data.columns.tolist())
```

## 6.Cleaning Text-Based Columns
## We standardize all string columns by removing extra spaces and converting inconsistent missing-value labels into proper NaN values.

```python
# Identify object (string) columns
text_cols = df_data.select_dtypes(include='object').columns
# Clean each text column
for col in text_cols:
    df_data[col] = df_data[col].astype(str).str.strip()  # remove spaces around
    df_data[col] = df_data[col].replace(
        ['', 'none', 'None', 'NaN', 'nan'], 
        np.nan
    )  # normalize missing values

print("Text columns cleaned.")
df_data.head(40)
```

## 7.Cleaning and Converting the Age Column
## We standardize age values by converting text like "forty" into numeric form and coercing invalid entries into NaN.

```python
# 1. Map the text 'forty' to the number 40 by converting everything to string and lowercasing it.
df_data['age'] = df_data['age'].astype(str).str.lower().str.replace('forty', '40')
# 2. Convert the entire column to numeric. 
# 'errors='coerce' turns non-numeric values (like 'nan' strings or other remaining text) into np.nan.
df_data['age'] = pd.to_numeric(df_data['age'], errors='coerce')
# 3. Convert to an integer type that supports missing values (<NA>)
df_data['age'] = df_data['age'].astype('Int64')
df_data.head(40)
```

## 8.Imputing Missing Age Values
## Missing ages are replaced using the median age, which is a robust choice for skewed or irregular numeric data. This ensures the age column is complete without being distorted by outliers.

```python
# 1. Calculate the median age from the non-missing values.
age_median = df_data['age'].median()
# 2. Fill (impute) the missing values with the median.
df_data['age'] = df_data['age'].fillna(age_median)
df_data.head(40)
```

## 9.Formatting Column Names
## We capitalize each column name to improve readability and make the dataset look cleaner

```python
df_data.columns = [col.capitalize() for col in df_data.columns]
print(df_data.columns.tolist())
```

## 10.Standardizing Patient Names
## We format patient names using title case so that each word begins with a capital letter.

```python
# For making name first letter capital
df_data['Patient_name'] = df_data['Patient_name'].str.strip().str.title()
df_data.head(40)
```

## 11.Imputing Missing Condition Values
## We fill missing entries in the Condition column using the mode—the most frequently occurring medical condition.

```python
# 1. Calculate the mode (most frequent value).
condition_mode = df_data['Condition'].mode()[0]
# 2. Fill (impute) the missing values with the mode.
df_data['Condition'] = df_data['Condition'].fillna(condition_mode)
df_data.head(40)
```

## 12.Standardizing Visit Date Formats
## We convert all visit dates into a consistent YYYY-MM-DD format using a simple parsing function.

```python
def fix_date(x):
    try:
        return pd.to_datetime(x).strftime('%Y-%m-%d')
    except:
        return x   # if pandas can't parse it, keep original

df_data['Visit_date'] = df_data['Visit_date'].apply(fix_date)
df_data.head(40)
```

## 13.Extracting Systolic and Diastolic Blood Pressure
## We split the blood pressure column into separate Systolic and Diastolic values to convert the combined string format into meaningful numeric features. Converting them to numeric types enables proper analysis and handling of missing or invalid entries.

```python
# If the column name is different, change it here
bp_col = 'Blood_pressure'

# Split into systolic and diastolic
df_data[['Systolic', 'Diastolic']] = df_data[bp_col].str.split('/', expand=True)

# Convert to numeric
df_data['Systolic'] = pd.to_numeric(df_data['Systolic'], errors='coerce')
df_data['Diastolic'] = pd.to_numeric(df_data['Diastolic'], errors='coerce')

df_data.head(40)
```

## 14.Imputing Missing Blood Pressure Values
## We fill missing systolic and diastolic readings using their respective median values, ensuring the blood pressure data remains complete and medically reasonable. Median imputation helps avoid distortion from extreme values or measurement errors.

```python

# Fill systolic NaN with median
df_data['Systolic'] = df_data['Systolic'].fillna(df_data['Systolic'].median())

# Fill diastolic NaN with median
df_data['Diastolic'] = df_data['Diastolic'].fillna(df_data['Diastolic'].median())
df_data.head(40)
```

## 15.Cleaning and Imputing Cholesterol Values
## We convert cholesterol readings into numeric form and replace missing values using the median.

```python
# CLEAN CHOLESTEROL COLUMN

# Convert to numeric
df_data['Cholesterol'] = pd.to_numeric(df_data['Cholesterol'], errors='coerce')

# Fill NaN with median value
df_data['Cholesterol'] = df_data['Cholesterol'].fillna(df_data['Cholesterol'].median())

df_data.head(40)
```

## 16.Cleaning and Validating Email Addresses
## We standardize email entries by converting them to lowercase, replacing invalid or blank values with NaN, and keeping only entries that contain both “@” and “.”. Missing or invalid emails are then filled with a consistent placeholder to maintain completeness.

```python
# 1) Normalize blanks to NaN
df_data['Email'] = df_data['Email'].replace(['', ' ', 'nan', 'NaN', 'None', 'NONE'], np.nan)

# 2) Convert all emails to lowercase
df_data['Email'] = df_data['Email'].str.lower()

# 3) Very simple email validator (no regex)
def simple_email(x):
    if pd.isna(x):
        return np.nan
    
    x = str(x).strip()
    
    # basic rule: must contain "@" AND "."
    if "@" in x and "." in x:
        return x
    else:
        return np.nan

df_data['Email'] = df_data['Email'].apply(simple_email)

# 4) Fill missing with placeholder
df_data['Email'] = df_data['Email'].fillna('unknown@missing.com')

df_data.head(40)
```

## 17.Standardizing and Validating Phone Numbers
## We clean phone numbers by removing symbols and spaces, keeping only valid 10-digit numeric entries.

```python
# 1) Replace blanks with NaN
df_data['Phone_number'] = df_data['Phone_number'].replace(['', ' ', 'nan', 'NaN', 'None', 'NONE'], np.nan)

# 2) Convert to string
df_data['Phone_number'] = df_data['Phone_number'].astype(str)

# 3) Remove '-' and spaces (very simple)
df_data['Phone_number'] = df_data['Phone_number'].str.replace('-', '')
df_data['Phone_number'] = df_data['Phone_number'].str.replace(' ', '')

# 4) Keep only valid 10-digit numbers, else NaN
df_data['Phone_number'] = df_data['Phone_number'].apply(
    lambda x: x if x.isdigit() and len(x) == 10 else np.nan
)

# 5) Fill missing
df_data['Phone_number'] = df_data['Phone_number'].fillna('unknown')

df_data.head(40)
```

# Step 4 — Statistical Analysis & Hypothesis Testing

##  1.  Define the Hypothesis
## We create two groups based on medical condition:

## Heart Disease patients

## Non–Heart Disease patients
## These groups will be used for the t-test to check whether their average cholesterol levels differ significantly.

```python
# Group 1: Patients with Heart Disease
heart = df_data[df_data['Condition'] == 'Heart Disease']['Cholesterol'].dropna()

# Group 2: Patients without Heart Disease
non_heart = df_data[df_data['Condition'] != 'Heart Disease']['Cholesterol'].dropna()

heart.head(), non_heart.head()
```

## 2. Perform Independent Two-Sample t-test
## This t-test compares the mean Cholesterol levels of the two patient groups.

```python
from scipy.stats import ttest_ind

# Independent t-test (unequal variances)
t_stat, p_value = ttest_ind(heart, non_heart, equal_var=False)

t_stat, p_value
```

## 3. Interpretation of Results
## We compare the mean cholesterol levels of both patient groups and use p < 0.05 as the significance threshold:
## If p < 0.05 → There is a statistically significant difference
## If p ≥ 0.05 → There is no statistically significant difference

```python
print("Mean Cholesterol (Heart Disease):", heart.mean())
print("Mean Cholesterol (Non-Heart Disease):", non_heart.mean())

if p_value < 0.05:
    print("\nConclusion: Reject H₀ → There is a significant difference in cholesterol levels.")
else:
    print("\nConclusion: Fail to reject H₀ → No significant difference in cholesterol levels.")
```

## 4. Confidence Interval for Cholesterol (Heart Disease Patients)
### This computes a 95% confidence interval for the mean cholesterol level of Heart Disease patients.
### It tells us the range in which the true average cholesterol for this group is likely to fall with 95% confidence.

```python
import scipy.stats as stats
import numpy as np

confidence_interval = stats.t.interval(
    0.95,                 # confidence level (first positional argument)
    len(heart) - 1,       # degrees of freedom
    loc=heart.mean(),     # mean
    scale=stats.sem(heart)  # standard error
)

confidence_interval
```

## 5. Type I and Type II Errors
## Type I Error:
### Rejecting a true null hypothesis.
### In this context: concluding that cholesterol levels are significantly different between Heart Disease and Non–Heart Disease patients when in reality, there is no real difference.

## Type II Error:
### Failing to reject a false null hypothesis.
### In this context: concluding that both groups have similar cholesterol levels when in reality, there is a meaningful difference that the test failed to detect.

# Step 5: Modeling and Pattern Discovery

# Step 1: Select Numeric Features
### We choose clinically important numerical variables for clustering.
### These features help group patients with similar health profiles

```python
# Step 1: Select numeric features for clustering
df_cluster = df_data[['Age', 'Systolic', 'Diastolic', 'Cholesterol']].copy()

df_cluster.head()
```

# Step 2: Handle Missing Values
### We fill missing values using median, which is appropriate for skewed biological data.

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
df_cluster_imputed = imputer.fit_transform(df_cluster)
```

# Step 3: Scale the Features
### K-Means is distance-based, so scaling is required.
### We use StandardScaler to bring all features to the same scale.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_cluster_scaled = scaler.fit_transform(df_cluster_imputed)
```

# Step 4:Find Optimal k (Elbow Method)
### We test different cluster numbers and find the best "elbow."

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []  # within-cluster sum of squares

for k in range(1, 10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(df_cluster_scaled)
    wcss.append(km.inertia_)

plt.plot(range(1, 10), wcss, marker='o')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.show()
```

# Step 5: Apply K-Means with optimal k
### Let’s assume the elbow suggests k = 3.

```python
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(df_cluster_scaled)

df_data['Cluster'] = clusters
df_data.head()
```
